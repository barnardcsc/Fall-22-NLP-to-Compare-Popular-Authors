```python
# file reading 
from urllib import request
import requests
result = requests.get('https://raw.githubusercontent.com/dipasur-star/NLP-workshop/main/book_1_grant.txt')
book_1 = result.text
print(book_1)
```

```python
# clean data 
# ask what we might do to only consider "important words"

# lower case
book_1 = book_1.lower()

print(book_1)
```

```python
# tokeniztion

book1_tokens = nltk.word_tokenize(book_1)
book1_tokens
```

```python
# remove punctuation

import string
# print(string.punctuation)
cleaned_book_1 = [w for w in book1_tokens if w not in string.punctuation]
cleaned_book_1
```

```python
# remove stop words

from nltk.corpus import stopwords
# print(stopwords.words('english'))

stopwords = stopwords.words('english')
# print(stopwords)
cleaned_book_1 = [w for w in cleaned_book_1 if w not in stopwords]
cleaned_book_1
```

```python
# remove other extraneous words
remove = ["''", "--", "``"]
cleaned_book_1 = [w for w in cleaned_book_1 if w not in remove]
cleaned_book_1
```

```python
# save into a function 
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
punctuation = string.punctuation
remove = ["''", "--", "``"]

def clean(tokens):
  cleaned = [w for w in tokens if w not in stopwords]
  cleaned = [w for w in cleaned if w not in punctuation]
  cleaned = [w for w in cleaned if w not in remove]
  return cleaned

# can call this function on tokenized text
```

```python
# lemmatization
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()

lemmatized_book_1 = [lemmatizer.lemmatize(word) for word in cleaned_book_1]

# test lemma sentence
lemma_sentence = " ".join(lemmatized_book_1)
print(lemma_sentence)
```

```python
# most common ngrams

# first for unigrams (bag of words), then tri-grams (more structured)

# try with plain text, cleaned tokens, and lemmas

# nltk.download('punkt')
from nltk.util import ngrams

# n = number of items in each ngram
n = 1
output = list(ngrams(cleaned_book_1, n))
frequency = nltk.FreqDist(output)

# top 20 most common ngrams
frequency.most_common(20)
```

```python
from transformers import RobertaTokenizerFast, TFRobertaForSequenceClassification, pipeline

tokenizer = RobertaTokenizerFast.from_pretrained("arpanghoshal/EmoRoBERTa")
model = TFRobertaForSequenceClassification.from_pretrained("arpanghoshal/EmoRoBERTa")

emotion = pipeline('sentiment-analysis', 
                    model='arpanghoshal/EmoRoBERTa', return_all_scores= True)

emotion_labels = emotion(lemma_sentence[:510])
emotion_labels
```

```python
emotions_sorted = sorted(emotion_labels[0], key=lambda d: d['score'], reverse = True)
emotions_sorted
```

```python
import matplotlib.pyplot as plt

label_score_list = [ ]
for element in emotions_sorted:
  label_score_list.append(list(element.values()))

label_score_dict = { }
for element in label_score_list:
  label_score_dict[element[0]] = element[1]

label_score_dict

# names = list(label_score_dict.keys())
# values = list(label_score_dict.values())
# plt.bar(range(len(label_score_dict)), values, tick_label=names)
# plt.show()
```
